# Databricks Connect — Getting Started

This notebook runs locally but all Spark operations execute on your Databricks workspace.

**Prerequisites:** Make sure your `.env` file is configured with `DATABRICKS_HOST` and `DATABRICKS_TOKEN`.

```python
from dbstarter import get_spark

spark = get_spark()
print("Connected!")
```

```python
# Verify the connection
spark.sql("SELECT 1 AS ok").show()
```

```python
# Read sample data
df = spark.read.table("samples.nyctaxi.trips")
df.show(5)
```

```python
# Run a SQL aggregation
spark.sql("""
    SELECT pickup_zip, COUNT(*) AS trips, ROUND(AVG(trip_distance), 2) AS avg_dist
    FROM samples.nyctaxi.trips
    GROUP BY pickup_zip
    ORDER BY trips DESC
""").show()
```

## Explore your workspace

Use the workspace utilities to browse catalogs, schemas, and tables programmatically.

```python
from dbstarter.workspace import list_catalogs, list_schemas, describe_table

# List all catalogs
for cat in list_catalogs():
    print(cat["name"])
```

```python
# Describe a table
info = describe_table("samples.nyctaxi.trips")
for col in info["columns"]:
    print(f"{col['name']:30s} {col['type']}")
```
